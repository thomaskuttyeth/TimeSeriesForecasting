## Linear Processes
A time series $X_t$ is defined to be linear process if it can be expressed as a linear combination of white noise as follows 

${X_t} = \mu + \sum\limits_{j =  - \infty }^\infty  {{\psi _j}{W_{t - j}}}$

where $\sum\limits_{j = - \infty }^\infty {\left| {{\psi _j}} \right|} < \infty$ and $ W_t \sim \mathcal{N}(0, \sigma^2)$

## Creating white noise

```python
import numpy as np 
import matplotlib.pyplot as plt

def make_white_noise(mean,std,n):
    return np.random.normal(mean,std,n)
mean = 0
std = 1 
num_samples = 150
samples = make_white_noise(mean,std,n = num_samples)
plt.plot(samples)
plt.title('White noise with mean 0 and std 1')
plt.show()
```

## Mean and 
Generally, the properties of a linear process related to mean and variance are given by 
$\begin{aligned}
\mu_{X} &= \mu \\
\gamma_{X}(h) &= \sigma _W^2\sum\limits_{j =  - \infty }^\infty  {{\psi _j}{\psi _{h + j}}}
\end{aligned}$

$\begin{aligned}
  \gamma \left( h \right) &= Cov\left( {{x_t},{x_{t + h}}} \right) \\
   &= Cov\left( {\mu  + \sum\limits_{j =  - \infty }^\infty  {{\psi _j}{w_{t - j}}} ,\mu  + \sum\limits_{j =  - \infty }^\infty  {{\psi _j}{w_{t + h - j}}} } \right) \\
   &= Cov\left( {\sum\limits_{j =  - \infty }^\infty  {{\psi _j}{w_{t - j}}} ,\sum\limits_{j =  - \infty }^\infty  {{\psi _j}{w_{t + h - j}}} } \right) \\
   &= \sum\limits_{j =  - \infty }^\infty  {{\psi _j}{\psi _{j + h}}Cov\left( {{w_{t - j}},{w_{t - j}}} \right)}  \\
   &= \sigma _w^2\sum\limits_{j =  - \infty }^\infty  {{\psi _j}{\psi _{j + h}}}  \\ 
\end{aligned}$

$ Cov\left( {{w_{t - j}},{w_{t + h - j}}} \right) = 0$ if $t - j \ne t + h - j$

## Backshit operator 
$B^k \, X_t = X_{t-k}$

So, the linear process can be rewrite in terms of backshift operator 

${X_t} = \mu + \psi \left( B \right){W_t}$
